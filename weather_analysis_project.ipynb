```python
# reading a csv file to alist of dictionaries

data = csv.DictReader(open("weather_data.csv"))
data = list(data)
print(data)
```

```python
# Extract temprature
temperature_values = []
for row in data:
  temperature_value = row['Temperature']
  if temperature_value != '':
     temperature_values.append(int(temperature_value))
print(temperature_values)
```

```python
#calculating the mean
temp_mean = sum(temperature_values)/len(temperature_values)

print(temp_mean)
```

```python
# rounding temp mean to zero decimal point
temp_mean = round(temp_mean)
print(temp_mean)
```

```python
#  Start from number, format for display
temp_mean = 51.199
formatted_mean = "{:.2f}".format(temp_mean)
print(formatted_mean)
```

```python
# updating null temp records with the temp_mean and converting str values to int

for row in data:
    if row['Temperature'] == '':
        row['Temperature'] = temp_mean
    row['Temperature'] = int(row['Temperature'])

print(data)
```

```python
# using list comprehension to create a list of temp values

temp_values = [row['Temperature'] for row in data]
print(temp_values)
```

```python
# getting max and min, count and mean temp values

max_temp = max(temp_values)
min_temp = min(temp_values)
count_temp = len(temp_values)
mean_temp = sum(temp_values)/len(temp_values)
print(max_temp)
print(min_temp)
print(count_temp)
print(mean_temp)
```

```python
# Initialize empty dictionary to hold city data
city_data = {}

# Loop through each row in your dataset
for row in data:
    city = row['City']
    temp_str = row['Temperature']

    # Skip empty or missing temperature values
    if temp_str == '':
        continue

    try:
        temp = float(temp_str)
    except ValueError:
        continue  # skip if conversion fails

    # If the city isn't already in the dictionary, add it
    if city not in city_data:
        city_data[city] = {
            'temp_sum': temp,
            'count': 1
        }
    else:
        city_data[city]['temp_sum'] += temp
        city_data[city]['count'] += 1
```

```python
# Calculate mean temperature per city
city_summary = []

for city, values in city_data.items():
    mean_temp = values['temp_sum'] / values['count']
    city_summary.append({
        'City': city,
        'Total_Temp': round(values['temp_sum'], 2),
        'Readings': values['count'],
        'Mean_Temp': round(mean_temp, 2)
    })
```

```python
# Create a refined list from city_data
refined_data = []

for city, values in city_data.items():
    temp_sum = values['temp_sum']
    count = values['count']
    mean_temp = round(temp_sum / count, 2)

    refined_data.append({
        'city_name': city,
        'temp_sum': temp_sum,
        'count': count,
        'temp_mean': mean_temp
    })

# Print the result
print(refined_data)
```
